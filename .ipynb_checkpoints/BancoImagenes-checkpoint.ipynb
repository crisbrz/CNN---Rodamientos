# 2. STFT a señales:
Metodologia:
- Cortar señales a 10 giros del eje
- Aplicar STFT a señal cortada
- Plotear el spectrograma y guardar como imagen etiquetada.


#### Función que divide los set de datos según el número de datos deseado.

```python
import glob
def make_dataset(datos, pos_dato_enlista,num_samples,n_imagenes): #pos_dato_enlista 0, 1, 2 ej: H-A-1, H-A-2-H-A-3
    files = glob.glob(datos)
    files = np.sort(files)
    data = sio.loadmat(files[pos_dato_enlista])
    signal = data['Channel_1']
    signal = signal.reshape(-1)
    num_segments = np.floor(len(signal[:num_samples*n_imagenes])/num_samples) #Calculo el número de segmentos
    slices = np.split(signal[0:int(num_segments*num_samples)], num_samples) # dividir en el número de segmentos (mediciones, segmentos)
    segmented_data = np.array(slices)
    return segmented_data
```

```python
index_inc = [143356, 143365, 133124, 163845, 153594, 143361] #De parte del encoder
index_dec = [71682, 71684, 71682, 81921, 81917, 81919]
#Velocidad aumentando
H_A_1 = make_dataset('./Dataset/IncSpeed/Healthy/*.mat',0, int(index_inc[0]), 10)
H_A_2 = make_dataset('./Dataset/IncSpeed/Healthy/*.mat',1, int(index_inc[1]), 10)
H_A_3 = make_dataset('./Dataset/IncSpeed/Healthy/*.mat',2, int(index_inc[2]), 10)

I_A_1 = make_dataset('./Dataset/IncSpeed/IR_Fault/*.mat',0, int(index_inc[3]), 10)
I_A_2 = make_dataset('./Dataset/IncSpeed/IR_Fault/*.mat',1, int(index_inc[4]), 10)
I_A_3 = make_dataset('./Dataset/IncSpeed/IR_Fault/*.mat',2, int(index_inc[5]), 10)

#Velocidad disminuyendo
H_B_1 = make_dataset('./Dataset/DecSpeed/Healthy/*.mat',0, int(index_dec[0]), 10)
H_B_2 = make_dataset('./Dataset/DecSpeed/Healthy/*.mat',1, int(index_dec[1]), 10)
H_B_3 = make_dataset('./Dataset/DecSpeed/Healthy/*.mat',2, int(index_dec[2]), 10)

I_B_1 = make_dataset('./Dataset/DecSpeed/IR_Fault/*.mat',0, int(index_dec[3]), 10)
I_B_2 = make_dataset('./Dataset/DecSpeed/IR_Fault/*.mat',1, int(index_dec[4]), 10)
I_B_3 = make_dataset('./Dataset/DecSpeed/IR_Fault/*.mat',2, int(index_dec[5]), 10)
```

### Función que guarda los archivos como imagenes.

```python
#señal = H_A_X ejemplo, etiqueta, IR o Normal

def banco_imagenes(señal, etiqueta, ancho_ventana, fs, codigo):
    numero_imagen = 0
    for i in range(10):
        f, t, Zxx = signal.spectrogram(señal[:,i], fs, window=signal.get_window('hamming', ancho_ventana),
                                   nperseg=ancho_ventana, noverlap=ancho_ventana/2, scaling='density', mode='magnitude')
        plt.pcolormesh(t, f, Zxx, cmap = 'gray')
        plt.axis('off')
        if codigo == 'IR':
            plt.savefig('./Banco de imagenes/Falla/'+ str(etiqueta) + str(numero_imagen) +'.png')
        if codigo == 'Normal':
            plt.savefig('./Banco de imagenes/Normal/'+ str(etiqueta) + str(numero_imagen) +'.png')
        numero_imagen +=1
```

# Generación banco de imagenes

```python
#Incrementando velocidad - Funcionamiento normal
banco_imagenes(H_A_1,'H-A-1_var_', 16834, 200000, 'Normal')
banco_imagenes(H_A_2,'H-A-2_var_', 16834, 200000, 'Normal')
banco_imagenes(H_A_3,'H-A-3_var_', 16834, 200000, 'Normal')

#Incrementando velocidad - Falla pista interna.
banco_imagenes(I_A_1,'I-A-1_var_', 16834, 200000, 'IR')
banco_imagenes(I_A_2,'I-A-2_var_', 16834,  200000, 'IR')
banco_imagenes(I_A_3,'I-A-3_var_', 16834, 200000, 'IR')

#Disminuyendo velocidad - Funcionamiento normal
banco_imagenes(H_B_1,'H-B-1_var_', 16834, 200000, 'Normal')
banco_imagenes(H_B_2,'H-B-2_var_', 16834, 200000, 'Normal')
banco_imagenes(H_B_3,'H-B-3_var_', 16834, 200000, 'Normal')

#Disminuyendo velocidad - Falla pista interna
banco_imagenes(I_B_1,'I-B-1_var_', 16834, 200000, 'IR')
banco_imagenes(I_B_2,'I-B-2_var_', 16834, 200000, 'IR')
banco_imagenes(I_B_3,'I-B-3_var_', 16834, 200000, 'IR')
```
