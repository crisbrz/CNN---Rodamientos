### Funciones de ventana:

```python
from scipy import signal
from scipy.fft import fft, fftshift
import matplotlib.pyplot as plt
import numpy as np
```

```python
window = signal.windows.blackman(51)
plt.plot(window)
plt.title("Ventana Blackman")
plt.ylabel("Amplitud")
plt.xlabel("Muestra")
plt.savefig('images_informe/blackman_w.svg')
```

```python
plt.figure()
A = fft(window, 2048) / (len(window)/2.0)
freq = np.linspace(-0.5, 0.5, len(A))
response = np.abs(fftshift(A / abs(A).max()))
response = 20 * np.log10(np.maximum(response, 1e-10))
plt.plot(freq, response)
plt.axis([-0.5, 0.5, -120, 0])
plt.title("Respuesta en frecuencia de la ventana Blackman")
plt.ylabel("Magnitud Normalizada [dB]")
plt.xlabel("Frecuencia Normalizada [ciclos por muestra]")
plt.savefig('images_informe/blackman_f.svg')
```

```python
window = signal.windows.gaussian(51, std=7)
plt.plot(window)
plt.title(r"Ventana Gaussian ($\sigma$=7)")
plt.ylabel("Amplitud")
plt.xlabel("Muestra")
plt.savefig('images_informe/gaussian_w.svg')
```

```python
A = fft(window, 2048) / (len(window)/2.0)
freq = np.linspace(-0.5, 0.5, len(A))
response = 20 * np.log10(np.abs(fftshift(A / abs(A).max())))
plt.plot(freq, response)
plt.axis([-0.5, 0.5, -120, 0])
plt.title(r"Respuesta en frecuencia de ventana Gaussian ($\sigma$=7)")
plt.ylabel("Magnitud Normalizada [dB]")
plt.xlabel("Frecuencia Normalizada [ciclos por muestra]")
plt.savefig('images_informe/gaussian_f.svg')
```
